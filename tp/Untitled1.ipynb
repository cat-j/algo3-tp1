```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

complejidadTeorica = pd.DataFrame()

ies = range(1,31)
tiempos = [2**i for i in ies]
complejidad = [(2**i)*np.log(i) for i in ies]

df = pd.DataFrame({'Agentes': ies, 'Tiempo (milisegundos)': tiempos, 'Complejidad 2^i': complejidad})

df
```

```python
plt.clf()
df.plot(x='Agentes')
plt.ylabel('Tiempo (milisegundos)')
plt.show()
```
